```python
from tkinter import *
window=Tk()
window.geometry("290x380")
window.title("CALCULATOR")
window.resizable(0,0)
window.configure(background="#000000")


operator=None
first_no=None
second_no=None



def digit1(num1):
    text_now=label1["text"]
    new_text=text_now + str(num1)
    label1.config(text=new_text)
    
    
def clear():
    label1.config(text="")
    
def get_operator(op):
    global first_no, operator
    first_no=int(label1["text"])
    operator=op
    label1.config(text="")
    
def get_result():
    global first_no,operator
    second_no=int(label1["text"])
    
    
    if operator == "+":
        label1.config(text=str(first_no + second_no))
    elif operator == "-":
        label1.config(text=str(first_no - second_no))
    elif operator == "*":
        label1.config(text= str(first_no* second_no))
    else:
        if second_no == 0:
            label.config(text="error")
        else:
            label1.config(text = str(first_no / second_no))
    
    
    
    
    
    
    

label1=Label(window,text=" ",width=25,height=2,font=("arial",30),bg="#E0E0EE",anchor=W)
label1.grid(row=0,column=0)

but7=Button(window,text=7,bg="#505050",fg="white",width=7,height=3,font=("arial"),command=lambda:digit1(7))
but7.place(x=0,y=100)


but8=Button(window,text=8,bg="#505050",fg="white",width=7,height=3,font=("arial"),command=lambda:digit1(8))
but8.place(x=75,y=100)

but9=Button(window,text=9,bg="#505050",fg="white",width=7,height=3,font=("arial"),command=lambda:digit1(9))
but9.place(x=150,y=100)

but_div=Button(window,text="/",bg="#FF9500",width=6,height=3,font=("arial"),command=lambda: get_operator("/"))
but_div.place(x=225,y=100)

but4=Button(window,text=4,bg="#505050",fg="white",width=7,height=3,font=("arial"),command=lambda:digit1(4))
but4.place(x=0,y=170)

but5=Button(window,text=5,bg="#505050",fg="white",width=7,height=3,font=("arial"),command=lambda:digit1(5))
but5.place(x=75,y=170)

but6=Button(window,text=6,bg="#505050",fg="white",width=7,height=3,font=("arial"),command=lambda:digit1(6))
but6.place(x=150,y=170)

but_mul=Button(window,text="*",bg="#FF9500",width=6,height=3,font=("arial"),command=lambda: get_operator("*"))
but_mul.place(x=225,y=170)

but1=Button(window,text=1,bg="#505050",fg="white",width=7,height=3,font=("arial"),command=lambda:digit1(1))
but1.place(x=0,y=240)

but2=Button(window,text=2,bg="#505050",fg="white",width=7,height=3,font=("arial"),command=lambda:digit1(2))
but2.place(x=75,y=240)

but3=Button(window,text=3,bg="#505050",fg="white",width=7,height=3,font=("arial"),command=lambda:digit1(3))
but3.place(x=150,y=240)

but_sub=Button(window,text="-",bg="#FF9500",width=6,height=3,font=("arial"),command=lambda: get_operator("-"))
but_sub.place(x=225,y=240)

but_c=Button(window,text="C",bg="red4",fg="white",width=7,height=3,font=("arial"),command=lambda:clear())
but_c.place(x=0,y=310)

but0=Button(window,text=0,bg="#505050",fg="white",width=7,height=3,font=("arial"),command=lambda:digit1(0))
but0.place(x=75,y=310)

but_equal=Button(window,text="=",bg="#FF9500",width=7,height=3,font=("arial"),command=get_result)
but_equal.place(x=150,y=310)

but_add=Button(window,text="+",bg="#FF9500",width=6,height=3,font=("arial"),command=lambda: get_operator("+"))
but_add.place(x=225,y=310)



window.mainloop()
```
